## **The famous duck statue has been stolen from Fiftyville!**

The town of Fiftyville has called upon you to solve the mystery of the stolen duck. Authorities believe that the thief stole the duck and then, shortly afterwards, took a flight out of town with the help of an accomplice. Your goal is to identify:

- **Who the thief is,**
- **What city the thief escaped to, and**
- **Who the thief’s accomplice is who helped them escape.**

#### All you know is that the theft took place on **July 28, 2023**, and that it occurred on **Humphrey Street**.

### **How Will You Solve the Mystery?**

The Fiftyville authorities have collected some of the town’s records from around the time of the theft and prepared a SQLite database for you, `fiftyville.db`, which contains tables of data from various sources around the town. You can query that database using **SQL SELECT** queries to access the data of interest to you.

Using just the information in the database, your task is to solve the mystery.


### **Available Tables**:


- **crime_scene_reports**: Contains information about crime scenes, including the date and street location of the crime.
    - Columns: `id`, `year`, `month`, `day`, `street`, `description`
  
- **interviews**: Contains transcripts of interviews conducted with suspects or witnesses.
    - Columns: `id`, `name`, `year`, `month`, `day`, `transcript`

- **atm_transactions**: Records of ATM transactions made by residents of Fiftyville.
    - Columns: `id`, `account_number`, `year`, `month`, `day`, `atm_location`, `transaction_type`, `amount`
  
- **bank_accounts**: Information about bank accounts, including the account holder (person).
    - Columns: `account_number`, `person_id`, `creation_year`
  
- **airports**: Information about airports, including their abbreviation, full name, and city.
    - Columns: `id`, `abbreviation`, `full_name`, `city`
  
- **flights**: Details about flights, including the origin and destination airports, and the time of departure.
    - Columns: `id`, `origin_airport_id`, `destination_airport_id`, `year`, `month`, `day`, `hour`, `minute`
  
- **passengers**: Information about flight passengers and their seat assignments.
    - Columns: `flight_id`, `passport_number`, `seat`
  
- **phone_calls**: Records of phone calls made between individuals in Fiftyville.
    - Columns: `id`, `caller`, `receiver`, `year`, `month`, `day`, `duration`
  
- **people**: Personal information about the residents of Fiftyville, including their license plate numbers and phone numbers.
    - Columns: `id`, `name`, `phone_number`, `passport_number`, `license_plate`
  
- **bakery_security_logs**: Logs from the bakery security system, capturing vehicle activity and license plates.
    - Columns: `id`, `year`, `month`, `day`, `hour`, `minute`, `activity`, `license_plate`

# Analyze-ATM-Transactions


## Retrieve-ATM-Transaction-Data
Selected all records and filtered from the `atm_transactions` table to examine transactions made.


```python
%%sql
SELECT * 
FROM atm_transactions 
WHERE (year = 2021 AND month = 7 AND day = 28) AND atm_location = 'Leggett Street';
```

## Retrieve-Bank-Account-Data
Selected all records from the `bank_accounts` table to obtain `person_id` and `account_number` information.

```python
%%sql
Select * FROM bank_accounts
limit 5;
```

## Join-Both-Tables
Joined `atm_transactions` and `bank_accounts` tables using `account_number` as the key to identify the `person_id` involved in each transaction.


```python
%%sql
SELECT A.transaction_type,A.amount,B.person_id	
FROM atm_transactions AS A
JOIN bank_accounts AS B ON A.account_number = B.account_number
WHERE (year = 2021 AND month = 7 AND day = 28) AND atm_location = 'Leggett Street';
```

## Store-Results-in-DataFrame
Stored the joined results in a Pandas DataFrame called `clues_atm_transaction` for further analysis.

```python
# Jalankan query SQL dan simpan hasilnya ke dalam variabel
result = %sql SELECT A.transaction_type, A.amount, B.person_id \
              FROM atm_transactions AS A \
              JOIN bank_accounts AS B ON A.account_number = B.account_number \
              WHERE (year = 2021 AND month = 7 AND day = 28) AND atm_location = 'Leggett Street';

# Tukar hasil kepada Pandas DataFrame
clues_atm_transaction = result.DataFrame()

# Cek hasilnya
clues_atm_transaction
```

***Person ID of the suspect identified in the transaction related to the incident***

# Investigate Bakery Security Logs

## Filter Security Logs
Select data from `bakery_security_logs` and filter based on clues from reports and interviews

```python
%%sql
SELECT *
FROM bakery_security_logs
WHERE (year = 2021 AND month = 7 AND day = 28) 
AND (hour = 10 AND minute BETWEEN 0 AND 30);
```

## Select Relevant People Data
Pick relevant entries in `people` for individuals identified in the filtered logs

```python
%%sql
SELECT *
FROM people limit 5;
```

## Join on License Plate
Join `bakery_security_logs` with `people` using `license_plate` to get each person's `passport_number`.

```python
%%sql 
SELECT 
    A.id,
    A.name,
    A.passport_number,
    A.license_plate,
    B.activity,
    B.hour,
    B.minute
FROM 
    people AS A 
JOIN 
    bakery_security_logs AS B 
ON 
    A.license_plate = B.license_plate
WHERE 
    (B.year = 2021 AND B.month = 7 AND B.day = 28) 
    AND (B.hour = 10 AND B.minute BETWEEN 0 AND 30);
```

Nama suspek yg terlihat kamera cctv di tempat kejadian

## Store and Format Results
Save the joined result as `clues_security_logs` and format it into a Pandas DataFrame for analysis.

```python
# Run the SQL query and save the results to a DataFrame
result = %sql SELECT \
    A.id, \
    A.name, \
    A.passport_number, \
    A.license_plate, \
    B.activity, \
    B.hour, \
    B.minute \
FROM \
    people AS A \
JOIN \
    bakery_security_logs AS B \
ON \
    A.license_plate = B.license_plate \
WHERE \
    (B.year = 2021 AND B.month = 7 AND B.day = 28) \
    AND (B.hour = 10 AND B.minute BETWEEN 0 AND 30);

# Convert the results to a Pandas DataFrame
clues_security_logs = result.DataFrame()

# Display the DataFrame
clues_security_logs
```

# Trace Flight Passengers

## Filter Flights
Select data from the `flights` table and filter it using clues from reports and interviews.

```python
%%sql
SELECT * FROM flights
WHERE year = 2021 AND month = 7 AND day = 29
order by hour;
```

***This is the suspect, so we will order the results by the earliest flight taken, which is at 8:20 AM***

## Get Passenger Information
Select data from the `passenger` table to find the `passport_number

```python
%sql SELECT * FROM passengers limit 5;
```

## Join Flights and Passengers
Join the `flights` and `passenger` tables using `id` and `flight_id` to get the `passport_number`.
Arrange the results by the earliest ticket booking time since the suspect booked their ticket first

```python
%%sql
Select A.origin_airport_id,A.destination_airport_id,A.hour,A.minute,B.passport_number,B.seat
FROM flights AS A
JOIN passengers AS B ON A.id=B.flight_id
WHERE year = 2021 AND month = 7 AND day = 29
order by hour limit 8;
```

## Store result
Save the results (DataFrame) of the join operation into a variable called `clues_fly`

```python
# Run the SQL query and save the results into a DataFrame
result = %sql SELECT A.origin_airport_id, A.destination_airport_id, A.hour, A.minute, B.passport_number, B.seat \
FROM flights AS A \
JOIN passengers AS B ON A.id = B.flight_id \
WHERE year = 2021 AND month = 7 AND day = 29 \
ORDER BY hour LIMIT 8;

# Convert the results to a Pandas DataFrame
clues_fly = result.DataFrame()

# Set display options to ensure large numbers are shown without scientific notation
pd.options.display.float_format = '{:,.0f}'.format

# Check the results
clues_fly
```

***This is list of suspect who take the earliest flight***

# Check Phone Records

## Filter Phone Calls
Select data from the `phone_calls` table and filter it using clues gathered from reports and interviews.

```python
%%sql
SELECT * FROM phone_calls 
WHERE year = 2021 AND month = 7 AND day = 28 AND duration < 60;
```

## Select Relevant People Data
After filtering the phone calls, select data from the `people` table to identify individuals involved

```python
%sql SELECT * FROM people limit 5;
```

## Join Phone Calls and People
- Perform an inner join between the `phone_calls` and `people` tables using both the `phone_number` and `caller` as common keys.
- This join will help retrieve relevant information about the individuals associated with the phone calls.

```python
%%sql
SELECT
A.id,A.name,A.phone_number,A.passport_number,A.license_plate,B.receiver
FROM people AS A
JOIN phone_calls AS B ON A.phone_number = B.caller
WHERE year = 2021 AND month = 7 AND day = 28 AND duration < 60;
```

## Store Results in DataFrame
Save the joined results in a variable called `clues_phone_calls` and convert it into a formal Pandas DataFrame for further analysis.

```python
# Run the SQL query and save the results into a DataFrame
result = %sql SELECT A.id, A.name, A.phone_number, A.passport_number, A.license_plate, B.receiver \
FROM people AS A \
JOIN phone_calls AS B ON A.phone_number = B.caller \
WHERE year = 2021 AND month = 7 AND day = 28 AND duration < 60;

# Convert the results to a Pandas DataFrame
clues_phone_calls = result.DataFrame()

# Display the DataFrame
clues_phone_calls
```
